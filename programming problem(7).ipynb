```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
x = np.random.uniform(0,5,200)
noise = np.random.normal(0,1,200)
y = 3 + 4*x + noise
X = np.c_[np.ones(len(x)), x]
theta_cf = np.linalg.inv(X.T @ X) @ X.T @ y
print("Closed Form:", theta_cf)
theta = np.zeros(2)
eta = 0.05
m = len(y)
loss = []
for i in range(1000):
    preds = X @ theta
    error = preds - y
    grad = (2/m) * X.T @ error
    theta = theta - eta * grad
    loss.append(np.mean(error**2))
print("Gradient Descent:", theta)
plt.scatter(x,y,label="Data")
plt.plot(x, X@theta_cf, label="Closed Form")
plt.plot(x, X@theta, label="Gradient Descent")
plt.legend()
plt.show()
plt.plot(loss)
plt.title("Loss Curve")
plt.show()
```
